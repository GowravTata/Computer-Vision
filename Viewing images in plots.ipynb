```python
import  matplotlib.pyplot as plt
# Path is used to store the location in a variable 
from pathlib import Path
```

```python
data = Path('F:/Data/Swim/')
train = data/'train/images'
test = data/'test/images'
```

```python
# Usually we can view any image using simple code, suppose we want to view multiple images this can be handy
def show_images(images, PATH):
    for i in images:
        fig,ax = plt.subplots(1,1)
        img = plt.imread(PATH/i)
        plt.imshow(img)
```

```python
show_images(['000000001.jpg','000000002.jpg','000000003.jpg','000000004.jpg'],test)
```
